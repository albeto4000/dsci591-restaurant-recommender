```python
import pandas as pd
import plotly.express as px
import plotly.io as pio
pio.renderers.default = 'png'
```

```python
users = pd.read_csv('data/users.csv')
with pd.option_context('display.max_columns', None):
    display(users.head())
```

```python
restaurants = pd.read_csv('data/restaurants.csv')
display(restaurants.head())
```

```python
ratings = pd.read_csv('data/reviews.csv')

display(ratings.head())
```

```python
n_restaurants = len(restaurants)
n_ratings = len(ratings)
n_users = ratings['user_id'].nunique()

print(f'There are {n_ratings} ratings for {n_restaurants} restaurants, created by {n_users} unique users')
```

```python
ratings
```

```python
density = n_ratings / (n_users * n_restaurants)

print(f'The density of the dataset is {density:.6f}')
```

```python
ratings_per_user = ratings['user_id'].value_counts().reset_index()

fig = px.ecdf(ratings_per_user, x = 'count', template = 'plotly_white', log_x = True)

fig.update_layout(
    xaxis = dict(
    tickmode = 'array',
    tickvals = [0] + [10 ** i for i in range(5)],
    ticktext = ['0'] + ['10<sup>' + str(i) +'</sup>' for i in range(6)],
    ),
    xaxis_title = 'Restaurants Rated',
    yaxis_title = '% of Users'
)


fig.show()
```

```python
ratings_counts = ratings['stars'].value_counts().reset_index()

fig = px.bar(ratings_counts, x = 'stars', y = 'count', template = 'plotly_white')
fig.show()

print(f'Median Rating: {ratings['stars'].median():.4f}')
print(f'Mean Rating: {ratings['stars'].mean():.4f}')
```

```python
res_stats = ratings.groupby('business_id')['user_id'].count().to_frame('n_ratings')

fig = px.ecdf(res_stats, x = 'n_ratings', template = 'plotly_white', log_x = True)

fig.update_layout(
    xaxis = dict(
    tickmode = 'array',
    tickvals = [0] + [10 ** i for i in range(5)],
    ticktext = ['0'] + ['10<sup>' + str(i) +'</sup>' for i in range(6)],
    ),
    xaxis_title = '# of Reviews',
    yaxis_title = '% of Restaurants'
)


fig.show()
```

```python
res_stats = pd.merge(res_stats, restaurants[['business_id', 'name', 'city', 'state', 'attributes', 'categories']], on = 'business_id')

with pd.option_context('display.max_columns', None):
    display(res_stats.nlargest(10, 'n_ratings')[['name', 'city', 'n_ratings', 'attributes', 'categories']])
```
