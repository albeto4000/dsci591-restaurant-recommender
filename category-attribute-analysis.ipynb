```python
restaurants = pd.read_parquet('data/restaurants.parquet')
with pd.option_context('display.max_columns', None):
    display(restaurants.head())
```

```python
ratings = pd.read_parquet('data/reviews.parquet')

display(ratings.head())
```

```python
#Imports
import ast
import pandas as pd
import plotly.express as px
import plotly.io as pio
pio.renderers.default = 'colab'
```

```python
#Cell 1: per-restaurant rating summary-
res_avg = (
    ratings.groupby('business_id')['stars']
    .agg(avg_stars='mean', n_reviews='count')
    .reset_index()
)
res_avg.head()
```

```python
#Cell 2: parse & explode categories
cats = restaurants[['business_id', 'categories']].dropna(subset=['categories']).copy()
cats['category'] = cats['categories'].str.split(', ')
cats = cats.explode('category')
cats['category'] = cats['category'].str.strip()

# Every restaurant carries the 'Restaurants' tag by construction — drop it as noise
cats = cats[cats['category'] != 'Restaurants'][['business_id', 'category']]
cats.head()
```

```python
# Cell 3: most common categories
cat_counts = cats['category'].value_counts().reset_index()
cat_counts.columns = ['category', 'n_restaurants']

top20 = cat_counts.head(20).sort_values('n_restaurants')
fig = px.bar(top20, x='n_restaurants', y='category', orientation='h',
             template='plotly_white', title='Top 20 restaurant categories')
fig.update_layout(yaxis_title='', xaxis_title='# of restaurants')
fig.show()
```

```python
# Cell 4: categories vs ratings
# Average the per-restaurant averages (so mega-restaurants don't dominate),
# and keep only categories with >= 50 restaurants to avoid noisy niche tags.
cat_ratings = cats.merge(res_avg, on='business_id')

cat_summary = (
    cat_ratings.groupby('category')
    .agg(n_restaurants=('business_id', 'nunique'),
         avg_stars=('avg_stars', 'mean'),
         total_reviews=('n_reviews', 'sum'))
    .reset_index()
)
cat_summary = cat_summary[cat_summary['n_restaurants'] >= 50]

print('Highest-rated categories:')
display(cat_summary.sort_values('avg_stars', ascending=False).head(10))
print('Lowest-rated categories:')
display(cat_summary.sort_values('avg_stars').head(10))
```

```python
#Cell 5: parse & flatten attributes into a clean feature table
def clean_value(v):
    """Turn a messy Yelp attribute string into a real Python value."""
    if not isinstance(v, str):
        return v
    try:
        return ast.literal_eval(v)       # 'True'->True, "'full_bar'"->'full_bar', dict-strings->dict
    except (ValueError, SyntaxError):
        return v.strip("'\"")            # fallback: strip stray quotes

def flatten_attrs(raw):
    """One restaurant's attributes (dict, dict-string, or None) -> flat {col: value}."""
    if raw is None:
        return {}
    d = raw if isinstance(raw, dict) else clean_value(raw)
    if not isinstance(d, dict):
        return {}
    out = {}
    for k, v in d.items():
        cv = clean_value(v)
        if isinstance(cv, dict):             # nested (e.g. Ambience, BusinessParking)
            for sk, sv in cv.items():
                out[f'{k}_{sk}'] = clean_value(sv)
        else:
            out[k] = cv
    return out

attr_df = pd.json_normalize(restaurants['attributes'].apply(flatten_attrs))
attr_df.insert(0, 'business_id', restaurants['business_id'].values)

print(attr_df.shape)
attr_df.head()
```

```python
# Cell 6: most common attributes by coverage
# 'coverage' = fraction of restaurants that have a non-null value for that attribute.
coverage = (attr_df.drop(columns='business_id').notna().mean()
            .sort_values(ascending=False).reset_index())
coverage.columns = ['attribute', 'coverage']

top20 = coverage.head(20).sort_values('coverage')
fig = px.bar(top20, x='coverage', y='attribute', orientation='h',
             template='plotly_white', title='Top 20 attributes by coverage')
fig.update_layout(yaxis_title='', xaxis_title='% of restaurants with this attribute',
                  xaxis_tickformat='.0%')
fig.show()
```

```python
# Cell 7: boolean attributes vs ratings
res_attr = attr_df.merge(res_avg, on='business_id')
bool_cols = [c for c in attr_df.columns
             if c != 'business_id' and attr_df[c].dropna().isin([True, False]).all()
             and attr_df[c].notna().any()]

rows = []
for c in bool_cols:
    sub = res_attr[[c, 'avg_stars']].dropna()
    if sub[c].nunique() < 2:
        continue
    g = sub.groupby(c)['avg_stars'].mean()
    vc = sub[c].value_counts()
    rows.append({'attribute': c,
                 'avg_if_true': g.get(True), 'avg_if_false': g.get(False),
                 'diff': g.get(True, float('nan')) - g.get(False, float('nan')),
                 'n_true': int(vc.get(True, 0)), 'n_false': int(vc.get(False, 0))})

attr_effect = pd.DataFrame(rows)
# keep only attributes with enough restaurants on each side to be meaningful
attr_effect = attr_effect[(attr_effect['n_true'] >= 30) & (attr_effect['n_false'] >= 30)].sort_values('diff', ascending=False)
attr_effect   # top = rating tends to be higher with this; bottom = lower
```

```python
# Cell 8: categorical attributes vs ratings
cat_attrs = ['Alcohol', 'NoiseLevel', 'WiFi', 'RestaurantsPriceRange2', 'RestaurantsAttire']
for c in cat_attrs:
    if c not in res_attr.columns:
        continue
    summary = (res_attr.dropna(subset=[c])
               .groupby(c)['avg_stars'].agg(avg_stars='mean', n='count')
               .reset_index().sort_values('avg_stars', ascending=False))
    print(f'\n=== {c} ===')
    display(summary)
```

```python
# Cell 9: save reusable analysis outputs
# Feature table for the recommender: one row per restaurant + parsed attribute columns
attr_df.to_parquet('attributes_clean.parquet', index=False)
# Supporting analysis tables
cat_summary.to_parquet('category_ratings.parquet', index=False)
attr_effect.to_parquet('attribute_effects.parquet', index=False)
print('Saved feature table:', attr_df.shape[0], 'restaurants x', attr_df.shape[1]-1, 'attribute columns')
print('Files written: attributes_clean.parquet, category_ratings.parquet, attribute_effects.parquet')
```
