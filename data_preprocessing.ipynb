```python
import pandas as pd
import json
```

```python
#Reads the JSON businesses file into a Pandas dataframe
business_df = pd.read_json("data/yelp_academic_dataset_business.json", lines = True)

#Selects all businesses with category 'Restaurants'
restaurants = business_df[business_df['categories'].str.contains('Restaurants')]

#Writes the restaurants data to a CSV
restaurants.to_parquet('data/restaurants.parquet', index = False)
```

```python
#Reads the review JSON file 10,000 lines at a time 
review_df = pd.read_json("data/yelp_academic_dataset_review.json", lines = True, chunksize=10000)

reviews = review_df[review_df['business_id'].isin(restaurants['business_id'])]

reviews.to_parquet('data/reviews.parquet', index = False)
```

```python
review_users = reviews['user_id'].unique()

#Reads the review JSON file 10,000 lines at a time 
user_df = pd.read_json("data/yelp_academic_dataset_user.json", lines = True, chunksize=10000)

datatypes = {
    'user_id': str,
    'name': str,
    'review_count': int,
    'yelping_since': str,
    'useful': int,
    'funny': int,
    'cool': int,
    'elite': str,
    'friends': str,
    'fans': int,
    'average_stars': float,
    'compliment_hot': int,
    'compliment_more': int
}

user_df = pd.read_csv('data/users.csv', dtype = datatypes)

users = user_df[user_df['user_id'].isin(review_users)]

users.to_parquet('data/users.parquet', index = False)
```
